```python
import keras
import numpy as np
import matplotlib.pyplot as pp
```

```python
X=[]
Y=[]
for i in range(600):
    v = np.random.random()*10
    X.append(v)
    if((int(v)%2)==0):
        Y.append(-1)
    else:
        Y.append(1)
pp.plot(X,Y,"ro")
```

```python
np.random.random()
```

```python
model=keras.models.Sequential()
model.add(layer=keras.layers.Dense(1000,input_shape=(1,)))
model.add(layer=keras.layers.Dense(30,activation='relu'))
model.add(layer=keras.layers.Dense(400,activation='relu'))
model.add(layer=keras.layers.Dense(30,activation='relu'))
model.add(layer=keras.layers.Dense(100,activation='relu'))
model.add(layer=keras.layers.Dense(500,activation='relu'))
model.add(layer=keras.layers.Dense(100,activation='tanh'))
model.add(layer=keras.layers.Dense(1400,activation='relu'))
model.add(layer=keras.layers.Dense(200,activation='relu'))
model.add(layer=keras.layers.Dense(500,activation='relu'))
model.add(layer=keras.layers.Dense(1))
```

```python
model.compile(loss="mse",optimizer='adam')
model.fit(X,Y,epochs=1000)
```

```python
q=np.arange(1,30,0.2)
Y_=model.predict(q)
pp.plot(q,Y_+0.1,"ro",color='b')
pp.plot(X,Y,"ro")
```
